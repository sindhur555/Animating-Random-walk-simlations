```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd
from random import randint
from matplotlib.animation import FuncAnimation
import sys
import ffmpeg
```

```python
# Defining the parameters of the SDE

mu = 0.1
sigma = 0.2
initial_value = 100.0
time_steps = 100
delta_t = 0.01
num_simulations = 1000
```

```python
rows = num_simulations
cols = 100
fig, ax = plt.subplots()
ax.set_xlim(0, cols)
ax.set_ylim(0, initial_value * 2)

X = np.full((rows, cols), None)

X[:, 0] = initial_value

def animate(i):
    
    if i == 0:
        pass
    else:
        X[:,i] = X[:,i-1] * (1 + mu * delta_t + sigma * np.sqrt(delta_t) * np.random.randn(num_simulations))
        
    ax.clear()  # Clear the previous plot
    for row in range(num_simulations):
        ax.plot(X[row, :i])  # Plot rows up to the ith column
    ax.set_xlabel('Time')
    ax.set_ylabel('Value')
    ax.set_title('Evolution of multiple simulations')
    
# run the animation
ani = FuncAnimation(fig, animate, frames=100, interval=250, repeat=False)

#plt.show()
    
from matplotlib import rc

# equivalent to rcParams['animation.html'] = 'html5'
rc('animation', html='html5')
ani
```
